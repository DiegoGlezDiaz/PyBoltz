```python
import sys
import warnings
import time
import numpy as np
warnings.filterwarnings("ignore", message="numpy.dtype size changed")
warnings.filterwarnings("ignore", message="numpy.ufunc size changed")
sys.path.append('../../src/Scripts/Python')

from Magboltz import Magboltz

obj = Magboltz()
```

```python
from IPython.utils import io
ENERGY=[]
EF =[]
N = [1,2,3,4,7]
with io.capture_output() as captured:
    for i in range(5):
        obj.__init__()
        obj.NGAS =1
        obj.NMAX = 1
        obj.IPEN = 0
        obj.ITHRM=1
        obj.EFINAL = 0.0
        obj.NGASN=[N[i],0,0,0,0,0]
        obj.FRAC=[100,0,0,0,0,0]
        obj.TEMPC = 23
        obj.NANISO = 2 #<<<<<<<< THIS IS IT
        obj.TORR = 750.062
        obj.EFIELD = 100
        obj.BMAG = 0
        obj.BTHETA = 90
        obj.Start()
        ENERGY.append(obj.ENERGY)
        EF.append(obj.EFINAL)
    
```

```python
s=["CF4","Ar","He4","He3","Xe"]
```

```python
summ =0
for i in range(4000):
    summ+=ENERGY[0][i]
print(summ)
```

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
plt.style.use('seaborn-white')
plt.figure(figsize=(20,10))
for i in range(5):
    x = ENERGY[i]
    keys = []
    for j in range(4000):
        keys.append(j*EF[i]/4000)
    kwargs = dict(histtype='stepfilled', alpha=0.5, ec="k")
    plt.hist(list(keys), weights=list(x),bins =4000,label=s[i],**kwargs)
    plt.xlabel('Energy of collision [Ev]',fontsize=24)
    plt.ylabel('Frequency [#]',fontsize=24)
    plt.legend(fontsize=24)
    plt.grid()
plt.tick_params(labelsize = 18)
plt.xlim(0,4)
plt.title("Energy distribution for 4E5 collisions",fontsize=24)
plt.show()
```
